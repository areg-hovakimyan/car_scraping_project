```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import re
```

```python
df = pd.read_csv('volkswagen_gti_2017-2020.csv')
```

```python
df.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
df.rename(columns = {'vehicles':'vehicle_count'}, inplace = True)
```

```python
df['avg_price'] = df['avg_price'].str.replace('$', '')
```

```python
def modify_dates():
    
    days = []
    for i in df['date_of_sale']:
        days.append(i[:2])
    df['day'] = days

    months = []
    for i in df['date_of_sale']:
        months.append(i[3:5])
    df['month'] = months
    
    years = []
    for i in df['date_of_sale']:
        years.append(i[6:])
    df['year'] = years
    df.drop('date_of_sale', axis = 'columns', inplace = True)
    df['day'].astype(dtype = int, copy = False)
    df['month'].astype(dtype = int, copy = False)
    df['year'].astype(dtype = int, copy = False)
modify_dates()
```
